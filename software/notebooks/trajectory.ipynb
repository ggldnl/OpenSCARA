# Trajectory

In this notebooks you can visualize different trajectories before using them on the actual robot.

```python
import matplotlib.pyplot as plt

def plot_points(points):
    """
    Plot the trajectory points using matlab.

    Args:
        points(List[tuple]): A list of (x, y) coordinates of the points.
    """

    # Extract x and y coordinates
    x_points, y_points = zip(*points)
    
    # Create the plot
    plt.figure(figsize=(6, 6))
    plt.plot(x_points, y_points, 'bo', label='Discretized Points')  # plot points
    plt.gca().set_aspect('equal', adjustable='box')

    # Add labels and legend
    plt.title('Points')
    plt.xlabel('X')
    plt.ylabel('Y')
    
    # Show the plot
    plt.grid(True)
    plt.show()

def format_points(points):
    for point in points:
        print(f"Point({point[0]:.2f}, {point[1]:.2f}, {point[2]:.2f})")
```

## Circle

```python
import numpy as np

def discretize_circle(radius, num_points, center=(0, 0)):
    """
    Discretizes a circle into a certain number of points.

    Args:
        radius (float): The radius of the circle.
        num_points (int): The number of points to discretize the circle into.
        center (tuple): The (x, y) coordinates of the circle's center.

    Returns:
        points (List[tuple]): A list of (x, y) coordinates of the points.
    """
    angles = np.linspace(0, 2 * np.pi, num_points, endpoint=False)
    points = [(center[0] + radius * np.cos(angle), 
               center[1] + radius * np.sin(angle)) for angle in angles]
    return points
```

```python
radius = 50
num_points = 8
center = (120, 0)
z = 100

circle_points = discretize_circle(radius, num_points, center)

# Add z coordinate
circle_points_z = [(point[0], point[1], z) for point in circle_points]

format_points(circle_points_z)
```

```python
plot_points(circle_points)
```

## Sinewave

```python
def generate_sine_wave(amplitude, frequency, num_points, x_range=(0, 2 * np.pi)):
    """
    Generates points for a sine wave.

    Args:
        amplitude (float): The amplitude of the sine wave.
        frequency (float): The frequency of the sine wave.
        num_points (int): The number of points to generate.
        x_range (tuple): The range of x values over which to generate the sine wave.

    Returns:
        points (List[tuple]): A list of (x, y) coordinates of the sine wave points.
    """
    x_values = np.linspace(x_range[0], x_range[1], num_points)
    y_values = amplitude * np.sin(frequency * x_values)
    
    points = list(zip(x_values, y_values))
    return points
```

```python
amplitude = 1
frequency = 2
num_points = 10
x_range = (0, np.pi)
x = 120

sine_wave_points = generate_sine_wave(amplitude, frequency, num_points, x_range)

# Add the x coordinate and change the order of the others such that the sinewave trajectory is not planar
sine_wave_points_z = [(x, point[0], point[1]) for point in sine_wave_points]

format_points(sine_wave_points_z)
```

```python
plot_points(sine_wave_points)
```
